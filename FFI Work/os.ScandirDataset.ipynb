```python
import os
import subprocess
from os import scandir
import time
```

```python
def listMaker(urlList):
    newUrlList = []
    
    for repository in range(0, len(urlList)-1):
        currUrl = urlList[repository]
        directory = currUrl.split("/")      
     
        try:
            newUrlList.append(directory[len(directory)-1])     #Stores name of the repository so it can be deleted later
        except:
            continue
    
    return newUrlList

def multCallsPerRepSearcher(rep, ffiList):
    foundItems = []
    foundKeywords = []
    
    for ffiString in ffiList:
        if ffiString in foundKeywords:
            continue
        else:
            process = subprocess.run(f'grep -rl \'{ffiString}\' {rep} | cut -f1 -d:', shell=True, stdout = subprocess.PIPE)
            output = process.stdout.decode('utf-8')
            
            if output != "":
                foundItems.append(ffiString+"^^^"+rep+"^^^"+output)
                foundKeywords.append(ffiString)
                if len(foundKeywords)>1:
                    return foundItems
                    
    return ["Nothing."]
        

def multCallsPerFileSearcher(file, ffiList):      
    foundItems = []            #List of file and grep output
    foundKeywords = []
        
    for ffiString in ffiList:
        process = subprocess.run(f'grep -n \'{ffiString}\' {file} | cut -f1 -d:', shell=True, stdout = subprocess.PIPE)
        output = process.stdout.decode('utf-8')
        
        #If there is a keyword match, append the file and line to grep list.
        if output != "":
            foundItems.append(ffiString+"^^^"+file+"^^^"+output)
            foundKeywords.append(ffiString)
            if len(foundKeywords) > 1:
                return foundItems
                
    return ["No presence of multiple foreign function calls."]

def subdirs(path):
    """Yield directory names not starting with '.' under given path."""
    folders = []
    for entry in os.scandir(path):
        if not entry.name.startswith('.') and entry.is_dir():
            folders.append(entry.name)
    return folders

def fileListMaker(scandir):
    fileList = []
    for i in scandir:
        if not i.name.startswith('.') and i.is_dir() and i != ".git" and i != ".gitignore":
            continue
        else:
            fileList.append(i.name)
    return fileList

def repWalker(path, ffiList):
    fileList = fileListMaker(scandir(path))
    for i in range(0, len(fileList)):
        fileList[i] = path+fileList[i]+"/"
    
    result = []
    for file in fileList:
        file = path+file+'/'
    
    result.extend(multCallsPerRepSearcher(fileList, ffiList))
    
    subfolders = subdirs(path)
    if len(subfolders) > 0:
        for folder in subfolders:
            result += repWalker(path+folder+"/", ffiList)

    return foundItems, foundKeywords

def fileWalker(path, ffiList):    
    fileList = fileListMaker(scandir(path))
    result = []
    
    for file in fileList:
        result.extend(multCallsPerFileSearcher(path+file, ffiList))
    
    subFolders = subdirs(path)
    if len(subFolders) > 0:
        for folder in subFolders:
            result += (fileWalker(path+folder+"/", ffiList))
    
    return result
# I suspect there's a problem here with directories. make sure path is correct.
```

```python
f = open("newFullList.csv", "r")
urlList = f.read()
f.close()
urlList = urlList.split(',')
#programDir = os.getcwd()
```

```python
ffiList = ["js.call(", "ctypes.DLL", "JS_CallFunctionName(", "cythonize(", "PyRun_SimpleString(", "PYBIND11_MODULE(", "FFI()", "SWIG", "jni.h", "JNICALL", "PyV8.JSContext()", "emscripten_run_script", "EM_JS("]
#Old ffiList = ["js.call(","ctypes.DLL","JS_CallFunctionName(","cythonize(", "PyRun_SimpleString(", "Py_initialize(", "pyObject", "PYBIND11_MODULE(", "FFI()", "JQuery.ajax(", "jni", "PyV8.JSContext()", "emscripten_run_script", "EM_JS("]
#Tester ffiList = ["tool", "both"]
repList = listMaker(urlList)
print(len(urlList))
for i in urlList:
    if " " in i:
        print(i)
```

```python
start_time = time.time()
result = {}
counter = 0
path = '/Users/dinobecaj/LocalDocuments/LyonsDB/'
for rep in range(0, len(repList)):
    repPath = path+repList[rep]
    print(f"{counter+1}. Analyzing {repList[rep]}...")
    
    foundItems = []            #List of file and grep output
    foundKeywords = []
        
    try:
        for ffiString in ffiList:
            process = subprocess.run(f'grep -ir \'{ffiString}\' {repPath} | cut -d: -f1 | uniq', shell=True, stdout = subprocess.PIPE)
            output = process.stdout.decode('utf-8')
        
            #If there is a keyword match, append the file and line to grep list.
            if output != "":
                foundItems.append(ffiString+"\n"+output)
                foundKeywords.append(ffiString)
    except:
        counter += 1
        result[repList[rep]] = ["Could not analyze."]
        continue
    
    counter += 1
    
    if len(foundKeywords) > 1:
        result[repList[rep]] = foundItems
    else:
        result[repList[rep]] = ["Repository does not contain multiple ffi's."]
    
print(result)
print("Program ran for:", (time.time()-start_time)/60, "minutes.")
```
