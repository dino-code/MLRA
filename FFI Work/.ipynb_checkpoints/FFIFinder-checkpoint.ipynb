```python
import os
import subprocess
```

```python
#Searches files for strings that are present when a foreign function is called.
def multCallsPerFileSearcher(filesList, ffiList):      
    for file in filesList:
        foundItems = []            #List of file and grep output
        foundKeywords = []
        
        for ffiString in ffiList:
            process = subprocess.run(f'grep -n \'{ffiString}\' {file} | cut -f1 -d:', shell=True, stdout = subprocess.PIPE)
            output = process.stdout.decode('utf-8')
        
            #If there is a keyword match, append the file and line to grep list.
            if output != "":
                foundItems.append(ffiString+"^^^"+file+"^^^"+output)
                foundKeywords.append(ffiString)
                if len(foundKeywords) > 1:
                    return foundItems
                
    return ["No files with multiple foreign function calls."]

def multCallsPerRepSearcher(filesList, ffiList):
    foundItems = []
    foundKeywords = []    
    
    for ffiString in ffiList:
        for file in filesList:
            if ffiString in foundKeywords:
                continue
            else:
                process = subprocess.run(f'grep -n \'{ffiString}\' {file} | cut -f1 -d:', shell=True, stdout = subprocess.PIPE)
                output = process.stdout.decode('utf-8')
            
                if output != "":
                    foundItems.append(ffiString+"^^^"+file+"^^^"+output)
                    foundKeywords.append(ffiString)
                    if len(foundKeywords) > 1:
                        return foundItems
    
    return ['No files with multiple foreign function calls.']
        
#Generates a list of all the files within a repository
def repFilesListGen():
    process = subprocess.run('git ls-files', shell=True, stdout = subprocess.PIPE)
    filesList = process.stdout.decode('utf-8')
    filesList = filesList.split('\n')
    
    return filesList

def writeToFilePerFile(multCallsPerFileDict):
    f = open('/Users/dinobecaj/Documents/ComputerScienceMS/LyonsWork/MLRA/FFI Work/ffCallsPerFile.csv', 'w')
    f.write('URL,')
    f.write('KEYWORD1,')
    f.write('FILE,')
    f.write('LINES,')
    f.write('KEYWORD2,')
    f.write('FILE,')
    f.write('LINES')
    f.write('\n')

    for key in multCallsPerFileDict:
        finds = multCallsPerFileDict[key]
        if finds != ['No files with multiple foreign function calls.']:
            key1 = finds[0]
            key2 = finds[1]
        
            splitValue1 = key1.split("^^^")
            splitValue2 = key2.split("^^^")
            f.write(key+",")
            f.write(splitValue1[0]+",")
            f.write(splitValue1[1]+",")
            f.write(splitValue1[2].replace('\n'," ")+",")
            f.write(splitValue2[0]+",")
            f.write(splitValue2[1]+",")
            f.write(splitValue2[2].replace('\n'," ")+"\n")
    
    f.close()
    
def writeToFilePerRep(multCallsPerRepDict):
    f = open('/Users/dinobecaj/Documents/ComputerScienceMS/LyonsWork/MLRA/FFI Work/ffCallsPerRep.csv', 'w')
    f.write('URL,')
    f.write('KEYWORD1,')
    f.write('FILE,')
    f.write('LINES,')
    f.write('KEYWORD2,')
    f.write('FILE,')
    f.write('LINES')
    f.write('\n')

    for key in multCallsPerRepDict:
        finds = multCallsPerRepDict[key]
        if (finds != None) and (finds != ['No files with multiple foreign function calls.']):
            key1 = finds[0]
            key2 = finds[1]
        
            splitValue1 = key1.split("^^^")
            splitValue2 = key2.split("^^^")
            f.write(key+",")
            f.write(splitValue1[0]+",")
            f.write(splitValue1[1]+",")
            f.write(splitValue1[2].replace('\n'," ")+",")
            f.write(splitValue2[0]+",")
            f.write(splitValue2[1]+",")
            f.write(splitValue2[2].replace('\n'," ")+"\n")
    
    f.close()
```

```python
f = open("newFullList.csv", "r")
urlList = f.read()
f.close()
urlList = urlList.split(',')
#programDir = os.getcwd()
```

```python
ffiList = ["cythonize(", "Py_initialize(", "pyObject", "PYBIND11_MODULE(", "FFI()"]
#previous ffiList = ["cythonize(", "PYBIND11_MODULE(", "FFI()", "Python.h"]
#Tester ffiList = ['asynchronous', 'Copyright']
multCallsPerFileDict = {}
multCallsPerRepDict = {}
```

```python
programDir = "/Users/dinobecaj/Documents/ComputerScienceMS/LyonsWork/MLRA/FFI Work"    #saves the main directory we are working from

counter = 0
for repository in range(0, len(urlList)-1):
    currUrl = urlList[repository]
    print(repository+1, "Now investigating:", currUrl)
    directory = currUrl.split("/")      
     
    try:
        directory = directory[len(directory)-1]     #Stores name of the repository so it can be deleted later
    except:
        continue

    currUrl = currUrl + ".git"
    
    try:
        os.system(f"git clone {currUrl}")        #Clones repository to local machine for analysis 
        print("Downloading repository")
    except:
        continue
    #try using grep on the file. If it doesn't work, simply move on to the next file.

    try:
        os.chdir(directory)                         #Enters the repository
        print(os.getcwd())
        filesList = repFilesListGen()              
        multCallsPerRepDict[urlList[repository]] = multCallsPerRepSearcher(filesList, ffiList)
        #multCallsPerFileDict[urlList[repository]] = multCallsPerFileSearcher(filesList, ffiList)
    except:
        os.chdir(programDir)
        os.system("rm -r " + directory)
        counter +=1
        continue

    os.chdir(programDir)

    os.system("rm -r " + directory)
    counter += 1                           #counter keeps track of script progress

    print(f"Progress: {counter}/{len(urlList)}")
print("Done.")
```
