```python
import os
import subprocess
```

```python
#Searches files for strings that are present when a foreign function is called.
def grepFile(file, ffiList, grepList, url, ffiRepositories):
    for ffiString in ffiList:
        process = subprocess.run(f'grep \'{ffiString}\' {file}', shell=True, stdout = subprocess.PIPE)
        output = process.stdout.decode('utf-8')
        if output != "":
            print(output, ffiString)
            grepList.append(file+" "+ffiString)
            ffiRepositories[url] = file+" "+ffiString
            break
            
```

```python
f = open("newFullList.csv", "r")
urlList = f.read()
f.close()
urlList = urlList.split(',')
```

```python
ffiList = ['Cython', 'pybind', 'cffi', 'ctypes', 'Python.h']
```

```python
for url in range(0, len(urlList)-1):
    currUrl = urlList[url]
    print(url, "Now investigating:", currUrl)
    directory = currUrl.split("/") 
```

```python
#Add a progress bar

ffiRepositories = {}
grepList = []

for url in range(0, len(urlList)-1):
    currUrl = urlList[url]
    print(url, "Now investigating:", currUrl)
    directory = currUrl.split("/")           
    
    try:
        directory = directory[4]     #Stores name of the repository so it can be deleted later
    except:
        continue
    
    url = url + ".git"
    os.system("git clone {}".format(url))        #Clones repository to local machine for analysis
    
    programDir = os.getcwd()
    
    try:
        os.chdir(directory)
    except:
        continue
    
    process = subprocess.run('git ls-files', shell=True, stdout = subprocess.PIPE)
    filesList = process.stdout.decode('utf-8')
    
    filesList = filesList.split('\n')
    
    for file in filesList:
        grepFile(file, ffiList, grepList, url, ffiRepositories)
        if url in ffiRepositories:
            break

    os.chdir(programDir)
    try:
        os.system("rm -r " + directory)
    except:
        continue
    
    print("Done")
```

```python
f = open('/Users/dinobecaj/Documents/ComputerScienceMS/LyonsWork/MLRA/ffiRepositories.csv', 'w')
f.write('URL,')
f.write('FILE,')
f.write('KEYWORD')
f.write('\n')
i = 0
for key in ffiRepositories:
    if i!= 0:
        f.write(key)
        f.write(',')
        try:
            splitValue = ffiRepositories[key].split(" ")
        except:
            continue
        f.write(splitValue[0]+',')
        f.write(splitValue[1]+'\n')
    i += 1
f.close()
```
